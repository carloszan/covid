```python
import pandas as pd
import zipfile
import os
```

```python
zip_file_path = 'HIST_PAINEL_COVIDBR_05out2024.zip'
extract_to_dir = 'extraido'

# Ensure the extraction directory exists
os.makedirs(extract_to_dir, exist_ok=True)

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to_dir)
```

```python
import rarfile

rar_file_path = 'HIST_PAINEL_COVIDBR_12out2024.rar'
extract_to_dir = 'extraido'

# Ensure the extraction directory exists
os.makedirs(extract_to_dir, exist_ok=True)

with rarfile.RarFile(zip_file_path, 'r') as rf:
    rf.extractall(path=extract_to_dir)
```

```python
from sqlalchemy import create_engine

engine = create_engine('postgresql://root:dietpi@192.168.3.200:5432/covid')

def load_data(file):
  df = pd.read_csv(f'extraido/{file}', sep=';')
  print('file is in memory')
  df.to_sql('covid', con=engine, if_exists='append', index=False)

for file in files:
  print(f'Trying file: {file.name}')
  load_data(file.name)
  print('Done')
```
